```python
import pandas as pd
df = pd.read_csv("/content/dermatology_database_1.csv")
df
```

```python
import numpy as np

# Replace '?' with NaN
df['age'] = df['age'].replace('?', np.nan)

# Convert to numeric (will convert valid strings to numbers and keep NaNs)
df['age'] = pd.to_numeric(df['age'])

# Optionally fill or drop NaNs before converting to int
df['age'] = df['age'].fillna(0).astype(int)  # or .dropna() before astype
```

```python
df.info()
```

```python
x=df.drop("class",axis=1)
y=df["class"]
```

```python
x.shape,y.shape
```

```python
from sklearn.preprocessing
```

```python
x = df.drop('class', axis=1)
y = df['class'] - 1
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=42
)

```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    use_label_encoder=False,
    eval_metric='mlogloss'  # For multi-class classification
)

model.fit(x_train, y_train)
```

```python
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5],
    'learning_rate': [0.1, 0.01],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

model1 = XGBClassifier(
    use_label_encoder=False,
    eval_metric='mlogloss'
)

grid_search = GridSearchCV(
    estimator=model1,
    param_grid=param_grid,
    cv=5,                      # 5-fold cross-validation
    scoring='accuracy',       # or 'f1_macro' for imbalanced multi-class
    n_jobs=-1,                 # use all CPU cores
    verbose=1
)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_

y_pred = best_model.predict(x_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
results_df = pd.DataFrame(grid_search.cv_results_)
results_df[['params', 'mean_test_score', 'rank_test_score']].sort_values(by='rank_test_score')
```
